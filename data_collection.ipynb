```python
# Utility functions

def fetch_audio_features(track_ids):
    """Fetch audio features for a list of track IDs."""
    audio_features = {}
    for i in range(0, len(track_ids), 100):  # Batch processing
        batch = track_ids[i:i + 100]
        features = sp.audio_features(batch)
        for feature in features:
            if feature:
                audio_features[feature["id"]] = feature
        time.sleep(1)  # Avoid rate limits
    return audio_features

def merge_features(track_data, audio_features):
    """Merge track data with audio features."""
    merged_data = []
    for track in track_data:
        features = audio_features.get(track["track_id"], {})
        merged_data.append({
            **track,
            **features,
        })
    return merged_data

def save_to_csv(data, filename):
    """Save merged data to a CSV file."""
    print(f"Saving merged data to {filename}...")
    fieldnames = set(key for row in data for key in row.keys())
    
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)
```

```python
def fetch_low_pop_playlist_tracks(playlist_id, playlist_name, playlist_genre, playlist_subgenre, limit=100):
    """Fetch tracks from a playlist and enrich with playlist metadata."""
    offset = 0
    track_data = []

    while True:
        try:
            results = sp.playlist_tracks(playlist_id, offset=offset, limit=limit)
        except spotipy.exceptions.SpotifyException as e:
            print(f"Spotify API error: {e}")
            time.sleep(5)
            continue
        
        tracks = results['items']
        if not tracks:
            break

        for item in tracks:
            track = item['track']
            if not track or 'popularity' not in track:
                continue
            
            print(f"Track popularity: {track['popularity']}")

            if track["album"].get("release_date") and track["album"]["release_date"][:4].isdigit():
                if int(track["album"]["release_date"][:4]) < 1990:
                    continue

            if 10 < track['popularity'] <= POPULARITY_THRESHOLD:
                print(f"This one is added. Track popularity: {track['popularity']}")

                track_data.append({
                    "track_id": track["id"],
                    "track_name": track["name"],
                    "track_artist": ", ".join(artist["name"] for artist in track["artists"]),
                    "track_popularity": track["popularity"],
                    "track_album_id": track["album"]["id"],
                    "track_album_name": track["album"]["name"],
                    "track_album_release_date": track["album"]["release_date"],
                    "playlist_name": playlist_name,
                    "playlist_id": playlist_id,
                    "playlist_genre": playlist_genre,
                    "playlist_subgenre": playlist_subgenre,
                })
        
        offset += limit
        time.sleep(1)  # Avoid rate limits
        print(f"Length of track data: {len(track_data)}")

    return track_data
```

```python
def fetch_high_pop_playlist_tracks(playlist_id, playlist_name, playlist_genre, playlist_subgenre, limit=100):
    """Fetch tracks from a playlist and enrich with playlist metadata."""
    offset = 0
    track_data = []

    while True:
        try:
            results = sp.playlist_tracks(playlist_id, offset=offset, limit=limit)
        except spotipy.exceptions.SpotifyException as e:
            print(f"Spotify API error: {e}")
            time.sleep(5)
            continue
        
        tracks = results['items']
        if not tracks:
            break

        for item in tracks:
            track = item['track']
            if not track or 'popularity' not in track:
                continue
            
            print(f"Track popularity: {track['popularity']}")

            if track['popularity'] >= POPULARITY_THRESHOLD:
                print(f"This one is added. Track popularity: {track['popularity']}")

                track_data.append({
                    "track_id": track["id"],
                    "track_name": track["name"],
                    "track_artist": ", ".join(artist["name"] for artist in track["artists"]),
                    "track_popularity": track["popularity"],
                    "track_album_id": track["album"]["id"],
                    "track_album_name": track["album"]["name"],
                    "track_album_release_date": track["album"]["release_date"],
                    "playlist_name": playlist_name,
                    "playlist_id": playlist_id,
                    "playlist_genre": playlist_genre,
                    "playlist_subgenre": playlist_subgenre,
                })
        
        offset += limit
        time.sleep(1)  # Avoid rate limits
        print(f"Length of track data: {len(track_data)}")

    return track_data
```

```python
all_tracks = []
for playlist in genres_playlists:
    tracks = fetch_high_pop_playlist_tracks(playlist["id"], playlist["name"], playlist["genre"], playlist["subgenre"])
    all_tracks.extend(tracks)

print(f"Total high-popularity tracks fetched: {len(all_tracks)}")

track_ids = [track["track_id"] for track in all_tracks]
audio_features = fetch_audio_features(track_ids)

merged_data = merge_features(all_tracks, audio_features)
save_to_csv(merged_data, "high_popularity_spotify_data.csv")
```

```python
all_tracks = []
playlist_tracks = {}

for playlist in genres_playlists:
    tracks = fetch_low_pop_playlist_tracks(
        playlist["id"], 
        playlist["name"], 
        playlist["genre"], 
        playlist["subgenre"]
    )
    playlist_tracks[playlist["id"]] = tracks  # Store tracks for later processing

total_tracks = sum(len(tracks) for tracks in playlist_tracks.values())

genre_track_limits = {
    playlist_id: int((len(tracks) / total_tracks) * 3300)
    for playlist_id, tracks in playlist_tracks.items()
}

for playlist_id, tracks in playlist_tracks.items():
    limit = genre_track_limits[playlist_id]
    all_tracks.extend(tracks[:limit])  # Add only up to the limit for this playlist

all_tracks = all_tracks[:3300]

print(f"Total tracks collected: {len(all_tracks)}")

track_ids = [track["track_id"] for track in all_tracks]
audio_features = fetch_audio_features(track_ids)

merged_data = merge_features(all_tracks, audio_features)
save_to_csv(merged_data, "low_popularity_spotify_data.csv")
```
